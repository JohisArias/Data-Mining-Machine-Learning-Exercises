### (b) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column

```r
# Crear vectores para las columnas de la matriz
x <- sample(1:100, size=12, replace=TRUE)

# Creando matriz a partir de vectores
matriz <- matrix(x, nrow = 3, ncol = 4)
 
# Agregar nombres de 
rownames(matriz) <- c('row0', 'row1', 'row2')
colnames(matriz) <- c('col0', 'col1', 'col2', 'col3')
 
# Imprimir el contenido de la matriz
matriz
rowSums(matriz)
colSums(matriz)
```

### (c) Create a DataFrame with 3 columns: "Name", "Age", and "City". Add at least 5 rows of data to the DataFrame. Then, filter the DataFrame to only include rows where the person's age is greater than or equal to 30.

```r
# Create a DataFrame with 3 columns: "Name", "Age", and "City".
# Add at least 5 rows of data to the DataFrame.
Names <- c('Alicia', 'Mario', 'Fernando', 'Ricardo', 'Franklin')
Ages <- c(59, 83, 74, 27, 9)
Cities <- c('Barcelona', 'Lima', 'Cartagena', 'New York', 'Berlin')

datos <- data.frame('Name'=Names,'Age'=Ages,'City'=Cities)
datos

# Agrega un nuevo registro
new <- data.frame('Name'= "Pelusa", 'Age'=2, 'City'="Miami")
datos <-rbind(datos, new)

'Data Frame actualizado'
datos

# Filter to only include rows where the person's age is greater than or equal to 30
filtrados <- subset(datos, Age>=30)
'Data Frame filtrado:'
filtrados
```
